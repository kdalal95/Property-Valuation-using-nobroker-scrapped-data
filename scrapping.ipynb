```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
browser= webdriver.Chrome(r"C:\Users\Krishna Dalal\Python Aegis\Projects\NoBroker\chromedriver.exe")
browser.get("https://www.nobroker.in")
```

# Bombay

```python
# mumbai subarban
```

```python
urls=[]
containers= browser.find_elements_by_id("listResults")
for i in containers:
    classes= i.find_elements_by_class_name("card-header-title")
    for j in classes:
        link= j.find_elements_by_tag_name("a")
        urls.append(link[0].get_attribute("href"))
            
```

```python
len(urls)
```

```python
lst=[]
for url in urls:
    webpage = requests.get(url).text
    page_soup = BeautifulSoup(webpage,'lxml')
    
    name = page_soup.find("h1",{"class":"nb__s_YQN"})
    name= name.text.strip()
    
    price= page_soup.find_all("div",{"class":"nb__qetBg"})[0].contents[0]
    price= price.text.strip()
    price= price.replace("₹ ","")
    
    area= page_soup.find_all("div",{"class":"nb__1sFxw"})[0].contents[0]
    area= area.text.strip()
    
    bedroom= page_soup.find_all("div",{"class":"nb__3vD7l"})[0].contents[0]
    bedroom= bedroom.text.strip()
    bedroom= bedroom.replace(" Bedroom", "")
    
    bathroom= page_soup.find_all("div",{"class":"nb__3vD7l"})[2].contents[0]
    bathroom= bathroom.text.strip()
    bathroom= bathroom.replace(" Bathroom", "")
    
    balcony= page_soup.find_all("div",{"class":"nb__3vD7l"})[4].contents[0]
    balcony= balcony.text.strip()
    
    parking= page_soup.find_all("div",{"class":"nb__3vD7l"})[6].contents[0]
    parking= parking.text.strip()
    
    power= page_soup.find_all("div",{"class":"nb__3vD7l"})[7].contents[0]
    power= power.text.strip()
    
    age= page_soup.find_all("div",{"class":"nb__2xbus"})[0].contents[0]
    age= age.text.strip()
    
    flooring= page_soup.find_all("div",{"class":"nb__2xbus"})[3].contents[0]
    flooring= flooring.text.strip()
    
    furnishing= page_soup.find_all("div",{"class":"nb__2xbus"})[5].contents[0]
    furnishing= furnishing.text.strip()
    
    facing= page_soup.find_all("div",{"class":"nb__2xbus"})[6].contents[0]
    facing= facing.text.strip()
    
    floor= page_soup.find_all("div",{"class":"nb__2xbus"})[7].contents[0]
    floor= floor.text.strip()
    
    try:
        livability= page_soup.find_all("div",{"span":"nb__3lxrH"})
        livability= livability.text.strip()
    except:
        livability= np.nan
        
    try:    
        transit= page_soup.find_all("div",{"class":"nb__1ZnkW"})[1]
        transit= transit.text.strip()
    except:
        transit= np.nan
        
    data= {'name': name, 'price':price, 'area':area, 'bedrooms':bedroom, 'bathroom':bathroom, 'balcony':balcony, 
           'parking':parking, 'power_backup':power, 'bldg_age':age, 'flooring':flooring, 'furnishing':furnishing, 
           'facing':facing, 'floor':floor, 'livability_score':livability, 'transit_score':transit}
    
    lst.append(data)
    
    #print(balcony, parking, power)
```

```python
df= pd.DataFrame(lst)
```

```python
df
```

```python
df.to_csv('mumbai_1')
```
